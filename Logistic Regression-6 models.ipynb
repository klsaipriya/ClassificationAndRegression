```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
%matplotlib inline



#Reading dataset using pandas
df=pd.read_csv('iris.csv')

#Storing the species data to evenly split
Y=df['variety']

#Splitting the data into train and test set
training_data,test_data=train_test_split(df,test_size=0.10,stratify=Y)

#Feature data
X_train=training_data[['sepal.length']]
#Target data
Y_train=training_data['sepal.width']

#lab_enc = preprocessing.LabelEncoder()
#encoded = lab_enc.fit_transform(Y_train)
#model=LogisticRegression()
#model.fit(X_train,encoded)
##Here, 
## r,s represents weight1 and weight2
## c represents intercept
## L represents learning rate
c=0

#Learning Rate
L=0.01
mse_list=[]

#training for 100 steps
epochs=20  
n_samples=135
k=0
r=0
s=0

#Lambda which is used for regularization
l=2

# Performing Training using Gradient Descent 
for epoch in range(epochs):
    #To calculate the average of mean_squared_error for n samples
    sum=0
    for i in range(0,135,32):
        #Linear Equation
        y_pred=r*X_train['sepal.length'][i:i+32]+s*X_train['sepal.length'][i:i+32]+c
        reg_term=l*((r+s)**2)
        D_m=(-2/n_samples)*((X_train['sepal.length'][i:i+32]*(Y_train[i:i+32]-y_pred)).sum())#Partial derivative of MSE wrt r and s
        D_c=(-2/n_samples)*((Y_train[i:i+32]-y_pred).sum())#Partial derivative of MSE wrt c
        r = r - L * D_m  # Update m
        s = s - L * D_m  # Update m
        c = c - L * D_c  # Update c
        print(mean_squared_error(Y_train[i:i+32],y_pred)+reg_term)#Adding loss of each batch
        err=mean_squared_error(Y_train[i:i+32],y_pred)+reg_term
        mse_list.append((k+1,err))
        k=k+1
        

plt.plot(*zip(*mse_list))
plt.title("Model : SL VS SW")
plt.xlabel("Step")
plt.ylabel("Mean Square Error")
#Calculating Accuracy on Test dataset
X_test=test_data[['sepal.length']]
Y_test=test_data['sepal.width']

val_test=np.dot(X_test,r)+np.dot(X_test,s)+c
y_test_predicted = 1/(1+np.exp(-val_test))
y_test_predicted=y_test_predicted>0.5
y_test_predicted=np.array(y_test_predicted,dtype='int64')
print("Error = ",mean_squared_error(Y_test,y_test_predicted))
```
