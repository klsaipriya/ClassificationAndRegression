```python
#importing the necessary packages 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from itertools import combinations
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

%matplotlib notebook
```

```python
#loading the Iris Dataset through Scikit
iris_dataset = datasets.load_iris()
#x-variable holds the attributes(sepal width,sepal length,petal length,petal width)
#y-variable holds the classes
x = iris_dataset.data
y = iris_dataset.target
```

```python
#Splitting the data with even class division
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.10, stratify=y)
```

```python
#Test plots for Training set
feature_names = iris_dataset.feature_names
comb = list(combinations(list(range(4)), 2))
fig = plt.figure(figsize=[8, 10])
for i, v in enumerate(comb):
    ax = fig.add_subplot(3, 2, i + 1)
    ax.scatter(X_train[:, v[0]], X_train[:, v[1]], c=Y_train, edgecolor='k')
    ax.set_xlabel(feature_names[v[0]])
    ax.set_ylabel(feature_names[v[1]])
```

```python
#Test plots for Test set
feature_names = iris_dataset.feature_names
comb = list(combinations(list(range(4)), 2))
fig = plt.figure(figsize=[8, 10])
for i, v in enumerate(comb):
    ax = fig.add_subplot(3, 2, i + 1)
    ax.scatter(X_test[:, v[0]], X_test[:, v[1]], c=Y_test, edgecolor='k')
    ax.set_xlabel(feature_names[v[0]])
    ax.set_ylabel(feature_names[v[1]])
```
